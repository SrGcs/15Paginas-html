```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Data.csv')
```

```python
df.describe()
```

```python
sns.histplot(df.Codigo,color="b", bins=30,kde=True) 
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.countplot(df.Codigo)
plt.show()
```

```python
correlacion = df.corr()
sns.heatmap(correlacion,xticklabels=correlacion.columns,yticklabels=correlacion.columns,annot=True)
```

```python
sns.relplot(x='Precio' ,y='Precio',hue='Precio',data=df)
```
